```
# using pyspark to load data using the [latest api]
# inspect blob storage landing location ~ json files arriving constantly
# read files from mount point ~ storage
# load data into dataframe (table in-memory) spread across the cluster
df_bank = spark.read.json("dbfs:/mnt/landing/bank/*.json")
df_credit_card = spark.read.json("dbfs:/mnt/landing/credit_card/*.json")
df_stripe = spark.read.json("dbfs:/mnt/landing/stripe/*.json")
df_device = spark.read.json("dbfs:/mnt/landing/device/*.json")
```

### Benefits to Convert to Apache Parquet

> **1** - acept nested and complex structures  
> **2** - compression-efficient and schema encoding  
> **3** - allows less disk read and fast queries  
> **4** - less I/O overhead due the batch-read ability  
> **5** - increases the scan read  

```
# format = apache parquet
# store on processing zone
# best practices = spark only sees the processing zone
# area where data is ready to be processed
# usually on parquet or any other format for big data workloads
df_bank.write.mode("append").parquet("/mnt/processing/parquet/batch/bank")
df_credit_card.write.mode("append").parquet("/mnt/processing/parquet/batch/credit_card")
df_stripe.write.mode("append").parquet("/mnt/processing/parquet/batch/stripe")
df_device.write.mode("append").parquet("/mnt/processing/parquet/batch/device")
```

```
# landing zone is cleaned after data processing of apache spark
# if not possible, merge or reading differential based on partitions 
# the most common approach is to remove everything from landing after conversion
# that guarantees that you don't have to read full cycles of data into storage

# set write mode = complete, append and overwrite
# append = atomically add bew data into existing delta table
write_delta_mode = "append"

# set processing zone location = dbfs:/mnt/owshq/delta/bronze
delta_processing_store_zone = "dbfs:/mnt/owshq/delta/batch/bronze"

# write into delta format
df_bank.write.mode(write_delta_mode).format("delta").save(delta_processing_store_zone + "/bank/")
df_credit_card.write.mode(write_delta_mode).format("delta").save(delta_processing_store_zone + "/credit_card/")
df_stripe.write.mode(write_delta_mode).format("delta").save(delta_processing_store_zone + "/stripe/")
df_device.write.mode(write_delta_mode).format("delta").save(delta_processing_store_zone + "/device/")
```
