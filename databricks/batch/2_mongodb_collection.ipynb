```
# OwsHQ_MongoDB = 7.3 LTS (includes Apache Spark 3.0.1, Scala 2.12)

"""
https://mvnrepository.com/artifact/org.mongodb.spark/mongo-spark-connector_2.12/3.0.2
mongo-spark-connector_2.12-3.0.2.jar
"""
# mongodb://root:w2nyCRBZKj@45.55.97.53:27017/owshq?authSource=admin&directConnection=true
# spark.mongodb.output.uri mongodb://root:w2nyCRBZKj@45.55.97.53:27017/owshq?authSource=admin&readPreference=primaryPreferred
# spark.mongodb.input.uri mongodb://root:w2nyCRBZKj@45.55.97.53:27017/owshq?authSource=admin&readPreference=primaryPreferred


"""
spark.mongodb.input.collection user
spark.mongodb.output.uri mongodb://root:w2nyCRBZKj@45.55.97.53:27017/owshq?authSource=admin&directConnection=true
spark.mongodb.input.uri mongodb://root:w2nyCRBZKj@45.55.97.53:27017/owshq?authSource=admin&directConnection=true
packages org.mongodb.spark:mongo-spark-connector:10.0.3
"""
```

```
# met nosql store - mongodb
# load [user] collection into a dataframe
# connect using the dataframe api = com.mongodb.spark.sql.DefaultSource
# reading entire collection
df_user = spark.read.format("com.mongodb.spark.sql.DefaultSource").option("database", "owshq").option("collection", "user").load()
display(df_user.printSchema())
```

```
# display data 
# based on schema infer
# struct type for nested values
# creates an object
display(df_user)
```
