```
# explicit configuration to access at the notebook level
spark.conf.set("fs.azure.account.key.owshqblobstg.blob.core.windows.net", "lmKoi0C1eXYJ/10fOqh+4im0CT8wyw/MWbUqmNooiH6yH2mmNS1HtiZNMXDUvCEc0TRtC/3i2eat+AStq3oURg==")

# using the [wasb] protocol to test read process
# time taken = 2.06 minutes
# 4.5 GB json file

# note = accessing blob storage without dbfs (mount) points
# not recommended since you will have to change location every time
# using mount points you can mount in any storage ~ s3, gcs, wasb, abfss, s3a and use everywhere
df_dataset_review_wasb = spark.read.json("wasbs://bs-stg-files@owshqblobstg.blob.core.windows.net/reviews/yelp_academic_dataset_review_1.json")
display(df_dataset_review_wasb)
```

```
# adls2 (abfss) - fastest protocol to access data
# designed for big data workloads
# uses hierarchical namespace = atomic directory manipulation

# https://docs.microsoft.com/en-us/azure/storage/blobs/data-lake-storage-introduction
# https://docs.microsoft.com/en-us/azure/storage/blobs/data-lake-storage-namespace

# abfss protocol
# time taken = 19.66 seconds
# 4.5 GB json file
df_dataset_review_abfss = spark.read.json("abfss://owshqblobstg@stgowshq.dfs.core.windows.net/yelp_academic_dataset_review_1.json")
display(df_dataset_review_abfss)
```

```
# https://docs.microsoft.com/en-us/azure/databricks/security/credential-passthrough/adls-passthrough
# https://docs.microsoft.com/en-us/azure/storage/blobs/data-lake-storage-use-databricks-spark

# create principal user with access on the adls2
# pro-tip = sometimes the adls passthrough conflicts with libraries

# name = Luan Moreno Medeiros Maciel
# adls2 role assignments = Storage Blob Data Contributor

# azure data lake storage credential passthrough - enable credential passthrough for user-level data access = enabled
# set configurations to access adls2
configs = {
  "fs.azure.account.auth.type": "CustomAccessToken",
  "fs.azure.account.custom.token.provider.class": spark.conf.get("spark.databricks.passthrough.adls.gen2.tokenProviderClassName")
}

# mount point of adls2 storage
dbutils.fs.mount(
  source = "abfss://owshqstorage@stgowshq.dfs.core.windows.net/",
  mount_point = "/mnt/adls2_owshqstorage",
  extra_configs = configs)
```

```
# list resources using mount point
# using the passthrough capability configured at the cluster level

# enable at the cluster level if error appears = com.databricks.backend.daemon.data.client.adl.azurecredentialnotfoundexception: could not find adls gen2 token
# enable credential passthrough for user-level data access at the cluster level

display(dbutils.fs.ls("/mnt/adls2_owshqstorage"))
```

```
# unmount link to azure data lake storage [gen2]
dbutils.fs.unmount("/mnt/adls2_owshqstorage")
```
